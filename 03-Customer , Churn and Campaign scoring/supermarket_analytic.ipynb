```python
# Pie chart customer lifestage
cust_lifestage_counts = super_csv['CUST_LIFESTAGE'].value_counts().to_dict()
cust_lifestage_df = pd.DataFrame(cust_lifestage_counts.items(), columns=['CUSTOMER_LIFESTAGE', 'COUNT'])
print(cust_lifestage_df)
print('\n')
plt.figure(figsize=(8, 8))
plt.pie(cust_lifestage_df['COUNT'], labels=cust_lifestage_df['CUSTOMER_LIFESTAGE'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribution of Categories')
plt.show()
```

```python
# df[(df.val > 0.5) & (df.val2 == 1)]
yf_df = super_csv[(super_csv['CUST_LIFESTAGE']== 'YF')]
prod_code_counts_yf = yf_df['PROD_CODE'].value_counts().to_dict()
print(prod_code_counts_yf)
```

```python
#  Who is the most spender
average_spend = super_csv['SPEND'].mean()
super_csv_filter_spend = super_csv[(super_csv['SPEND']>= average_spend)]
print(f'rows = {super_csv_filter_spend.shape[0]}')
cust_lifestage_filter_spend_counts = super_csv_filter_spend['CUST_LIFESTAGE'].value_counts().to_dict()
print(cust_lifestage_filter_spend_counts)
```

```python
# What type of shop that yf spend the most amount
yf_basket_type = yf_df['BASKET_TYPE'].value_counts().to_dict()
yf_basket_type.pop('XX')
basket_type_yf_lst = list(yf_basket_type.values())
basket_type_name = ['Full shop', 'Top up', 'Small shop']

condition_full_shop = yf_df['BASKET_TYPE'] == 'Full Shop'
condition_top_up = yf_df['BASKET_TYPE'] == 'Top Up'
condition_small_shop = yf_df['BASKET_TYPE'] == 'Small Shop'

sum_of_spend_full_shop = yf_df.loc[condition_full_shop, 'SPEND'].sum()
sum_of_spend_top_up = yf_df.loc[condition_top_up, 'SPEND'].sum()
sum_of_spend_small_shop = yf_df.loc[condition_small_shop, 'SPEND'].sum()
spend_yf_lst = [sum_of_spend_full_shop, sum_of_spend_top_up, sum_of_spend_small_shop]
print(f'full shop spend: {sum_of_spend_full_shop}\ntop up spend: {sum_of_spend_top_up}\nsmall shop spend: {sum_of_spend_small_shop}\n')


yf_spend_basket_type_df = pd.DataFrame(list(zip(basket_type_name,basket_type_yf_lst,spend_yf_lst)), 
                                       columns=['BASKET_TYPE','BASKET_TYPE_COUNT', 'TOTAL_SPENDING'])

plt.figure(figsize=(10, 6))

# Create bar chart
yf_spend_basket_type_df['BASKET_TYPE_COUNT'].plot(kind='bar')

# Create line graph with larger marker size
yf_spend_basket_type_df['TOTAL_SPENDING'].plot(kind='line', marker='o', markersize=10, color='black')

# Annotate data points on the line graph with larger font size
for i, row in yf_spend_basket_type_df.iterrows():
    plt.annotate(row['TOTAL_SPENDING'], (i, row['TOTAL_SPENDING']), textcoords="offset points", xytext=(0,15), ha='center', fontsize=12)

# Set labels and title
plt.xlabel('Basket Type')
plt.ylabel('Values')
plt.title('Basket Type vs. Values')

# Rotate x-axis labels for better visibility (optional)
plt.xticks(range(len(yf_spend_basket_type_df['BASKET_TYPE'])), yf_spend_basket_type_df['BASKET_TYPE'])

# Add legend
plt.legend(['Basket Count', 'Total Spending'])

# Show the plot
plt.show()
```

```python
# Define frequency of spending the product with time period of user yf
import itertools
yf_df_2006 = yf_df[(yf_df['SHOP_DATE']>= 20060000)&(yf_df['SHOP_DATE']< 20070000)]
yf_df_2007 = yf_df[(yf_df['SHOP_DATE']>= 20070000)&(yf_df['SHOP_DATE']< 20080000)]
yf_df_2008 = yf_df[(yf_df['SHOP_DATE']>= 20080000 )]

yf_df_prod_2006_obj = yf_df_2006['PROD_CODE_30'].value_counts().to_dict()
yf_df_prod_2007_obj = yf_df_2007['PROD_CODE_30'].value_counts().to_dict()
yf_df_prod_2008_obj = yf_df_2008['PROD_CODE_30'].value_counts().to_dict()

print(f'2006: {dict(itertools.islice(yf_df_prod_2006_obj.items(), 3))}, total prod: {yf_df_2006.shape[0]}\n2007: {dict(itertools.islice(yf_df_prod_2007_obj.items(), 3))}, total prod: {yf_df_2007.shape[0]}\n2008: {dict(itertools.islice(yf_df_prod_2008_obj.items(), 3))}, total prod: {yf_df_2008.shape[0]}')
```

```python
years = [2006, 2007, 2008]
product_g00007 = [yf_df_prod_2006_obj['G00007']/yf_df_2006.shape[0]*100, 
                  yf_df_prod_2007_obj['G00007']/yf_df_2007.shape[0]*100, 
                  yf_df_prod_2008_obj['G00007']/yf_df_2008.shape[0]*100]
product_g00004 = [yf_df_prod_2006_obj['G00004']/yf_df_2006.shape[0]*100, 
                  yf_df_prod_2007_obj['G00004']/yf_df_2007.shape[0]*100, 
                  yf_df_prod_2008_obj['G00004']/yf_df_2008.shape[0]*100]
product_g00016 = [yf_df_prod_2006_obj['G00016']/yf_df_2006.shape[0]*100, 
                  yf_df_prod_2007_obj['G00016']/yf_df_2007.shape[0]*100, 
                  yf_df_prod_2008_obj['G00016']/yf_df_2008.shape[0]*100]

# Create a DataFrame
data = pd.DataFrame({'Year': years, 'Product G00007': product_g00007, 'Product G00004': product_g00004, 'Product G00016': product_g00016})

# Set the figure size
plt.figure(figsize=(10, 6))

# Create a stacked bar chart
bars_a = plt.bar(data['Year'], data['Product G00007'], label='Product G00007', color='r')
bars_b = plt.bar(data['Year'], data['Product G00004'], label='Product G00004', bottom=data['Product G00007'], color='g')
bars_c = plt.bar(data['Year'], data['Product G00016'], label='Product G00016', bottom=data['Product G00007'] + data['Product G00004'], color='b')

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Ratio')
plt.title('Product Ratios Over 3 Years')
plt.legend()

# Annotate the values for each product in each layer
for bars, product in zip([bars_a, bars_b, bars_c], ['Product A', 'Product B', 'Product C']):
    for bar in bars:
        height = bar.get_height()

# Show the plot
plt.show()
print(product_g00007)
# Conclusion : 3 products ratio are still be nearly the same for 3 years
```

```python
# 2006 product pattern

start_period_2006 = 20060000
end_period_2006 = 20060100
quantity_g00007_2006 = []

for i in range(1,13):
    start_period_2006 += 100
    end_period_2006 += 100
    
    yf_df_g00007_period = yf_df[(yf_df['PROD_CODE_30']== 'G00007' )&
                         (yf_df['SHOP_DATE'] > start_period_2006 )&
                         (yf_df['SHOP_DATE'] < end_period_2006 )]
    quantity_g00007_2006.append(yf_df_g00007_period['QUANTITY'].sum())

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Set the figure size
plt.figure(figsize=(10, 6))

# Create a line chart
plt.plot(months, quantity_g00007_2006, marker='o', linestyle='-', color='b')
plt.ylim(600, 1300)
# Set labels and title
plt.xlabel('Month')
plt.ylabel('Quantity')
plt.title('Product G00007 Pattern Over 12 Months - 2006')

# Rotate x-axis labels for better visibility (optional)
plt.xticks(rotation=45)

# Show the plot
plt.grid()
plt.show()
```
